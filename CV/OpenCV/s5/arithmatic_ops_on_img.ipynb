```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image 
%matplotlib inline
from IPython.display import Image
```

```python
img_bgr = cv2.imread("New_Zealand_Coast.jpg", cv2.IMREAD_COLOR)

img_rgb = cv2.cvtColor(img_bgr,cv2.COLOR_BGR2RGB)

# Image(filename="New_Zealand_Coast.jpg")

# plt.imshow(img_bgr)

plt.imshow(img_rgb);
```

```python
mask = np.ones(img_rgb.shape, dtype="uint8") * 50

brighter = cv2.add(img_rgb, mask)

darker = cv2.subtract(img_rgb, mask)

plt.figure(figsize=[18,5])
plt.subplot(131); plt.imshow(darker);   plt.title("Darker");
plt.subplot(132); plt.imshow(img_rgb);  plt.title("Original");
plt.subplot(133); plt.imshow(brighter); plt.title("Brighter");
```

```python
mask1 = np.ones(img_rgb.shape) * 1.4
# mask1 = np.ones(img_rgb.shape) * .8
mask2 = np.ones(img_rgb.shape) * 1.2

darker = np.uint8(cv2.divide(np.float64(img_rgb), mask1))
# darker   = np.uint8(cv2.multiply(np.float64(img_rgb), mask1))
brighter = np.uint8(cv2.multiply(np.float64(img_rgb), mask2))

plt.figure(figsize=[18,5])
plt.subplot(131); plt.imshow(darker);   plt.title("Lower Contrast");
plt.subplot(132); plt.imshow(img_rgb);  plt.title("Original");
plt.subplot(133); plt.imshow(brighter); plt.title("Higher Contrast withou clipping");
```

```python
mask1 = np.ones(img_rgb.shape) * 1.4
mask2 = np.ones(img_rgb.shape) * 1.2

darker = np.uint8(cv2.divide(np.float64(img_rgb), mask1))
brighter = np.uint8(np.clip(cv2.multiply(np.float64(img_rgb), mask2),0,255))

plt.figure(figsize=[18,5])
plt.subplot(131); plt.imshow(darker);   plt.title("Lower Contrast");
plt.subplot(132); plt.imshow(img_rgb);  plt.title("Original");
plt.subplot(133); plt.imshow(brighter); plt.title("Higher Contrast with clipping");
```
