```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image 
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize'] = (9.0, 9.0)
from IPython.display import Image
```

```python
img = cv2.imread("Apollo_11_Launch.jpg",cv2.IMREAD_COLOR)

b,g,r = cv2.split(img)
img = cv2.merge((b,g,r))
# img[:,:,::-1]
plt.imshow(img)
```

```python
img_line = img.copy()


cv2.line(img_line, (200,400), (400,100),(255,255,0),thickness=6, lineType=cv2.LINE_AA)

plt.imshow(img_line)
```

```python
img_circle = img.copy()

cv2.circle(img_circle, (600,600), 100, (0,125,255), thickness=5, lineType=cv2.LINE_AA);

plt.imshow(img_circle)
```

```python
img_rect = img.copy()

cv2.rectangle(img_rect,(470,50),(700,600),(125,65,125),thickness=10,lineType=cv2.LINE_4);

plt.imshow(img_rect)
```

```python
img_text = img.copy()

cv2.putText(img_text, 
            text= "Orion 11",
            org= (800,200), 
            fontFace= cv2.FONT_HERSHEY_DUPLEX,
            fontScale= 2,
            color= (0,0,60),
            thickness= 10,
            lineType= cv2.LINE_AA
            )

plt.imshow(img_text)
```

```python
all_in_img = img.copy()


cv2.line(all_in_img, 
        (780,220), 
        (1080,220),
        (255,255,0),
        thickness=6, 
        lineType=cv2.LINE_AA)
cv2.putText(all_in_img, 
            text= "Orion 11",
            org= (800,200), 
            fontFace= cv2.FONT_HERSHEY_DUPLEX,
            fontScale= 2,
            color= (0,0,60),
            thickness= 10,
            lineType= cv2.LINE_AA
            )
cv2.circle(all_in_img,
          center= (600,600), 
          radius= 100, 
          color= (255,25,0), 
          thickness=5, 
          lineType=cv2.LINE_AA)
cv2.rectangle(all_in_img, 
              pt1= (450,50),
              pt2= (750,650),
              color= (0,200,0),
              thickness= 10,
              lineType= cv2.LINE_4
              )               
                

plt.imshow(all_in_img)
```
